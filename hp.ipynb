```python
import pandas as pd
import numpy as np
from datetime import datetime
from dateutil.relativedelta import relativedelta
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('data.csv',encoding = 'ISO-8859-1')
```

```python
df = df[df["price"]!=0]
```

```python
curr_time = datetime.now()
curr_time = str(curr_time).split(" ")[0]
curr_time
```

```python
date_format = "%Y-%m-%d"
df['date']=df['date'].apply(lambda x:relativedelta( datetime.strptime(curr_time,date_format),datetime.strptime(x.split(" ")[0],date_format)).years)
```

```python
df.price.max(), df.price.min()
```

```python
cur_time = datetime.now()
df['yr_built']=df['yr_built'].apply(lambda x:cur_time.year-x)
```

```python
df.head()
```

```python
df.drop(['street','statezip','country','date','sqft_lot','sqft_above','sqft_basement','yr_renovated'],axis = 1, inplace=True)
```

```python
df.head()
```

```python
df
```

```python
categorical_columns = ['city']
label_encoder = LabelEncoder()

# Fit the label encoder and transform the labels to numerical values
for column in categorical_columns:
    df[column] = label_encoder.fit_transform(df[column])
```

```python
df
```

```python
x = df.drop(['price'],axis=1)
y = df['price']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=45)
```

```python
sc_x = StandardScaler()
x_train = sc_x.fit_transform(x_train)
x_test = sc_x.transform(x_test)
```

```python
x_train
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
Y_pred = model.predict(x_test)
# Calculate metrics
mae = mean_absolute_error(y_test, Y_pred)
mse = mean_squared_error(y_test, Y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, Y_pred)

# Print the results
print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("R-squared (R²):", r2)

y_test_pre=model.predict(x_test)
test_acc=r2_score(y_test,y_test_pre)
print(f'accuracy = {test_acc*100}%')
```
